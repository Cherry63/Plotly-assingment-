```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
df=pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/matches.csv")
df.head(10)

```

4-->**Luckiest venue for each team(n-teams count - plots for each team(how much played vs how much won))**

```python
fig=px.bar(df,x='winner',y='city',color='venue')
fig.show()
```

1-->**Best teams based on winning count(1 plot)**

Based on the above graph we can say that CSK is the best team based on the most winning rate

```python
fig=px.bar(df,x="winner",y="toss_decision",color="toss_winner")
fig.show()
```

2-->**Best player based on player of the match(1 plot)**

Based on the above graph we can say that       is the best player based on the player_of_match data

```python
fig = px.bar(df,x="player_of_match",y="season",color="toss_winner")
fig.show()
```
